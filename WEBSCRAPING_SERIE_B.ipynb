```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import html5lib
import lxml
```

```python
url = "https://www.cbf.com.br/futebol-brasileiro/competicoes/campeonato-brasileiro-serie-b"
```

```python
res = requests.get(url)
```

```python
content = res.content
```

```python
  soup = BeautifulSoup(content,"html.parser")
table = soup.find("table")
tabela_str = str(table)
df_tabela = pd.read_html(tabela_str)[0]
df_tabela
#table = soup.find('tbody')
```

```python
df_tabela[['Pos','Evolução','Clube']] = df_tabela['Posição'].str.split(' ', expand=True,n=2)
```

```python
df_tabela.drop(['Posição'], axis=1)
df_tabela_final = df_tabela[['Pos','Evolução','Clube','PTS','J','V','E','D','GP','GC','SG','CA','CV','Recentes','Próx']]
df_tabela_final
```

```python
urltime = 'https://www.cbf.com.br/futebol-brasileiro/times/campeonato-brasileiro-serie-b/2023'
restime = requests.get(urltime)
content_time = restime.content
```

```python
soup_time = BeautifulSoup(content_time, "html.parser")
times = soup_time.find_all('div',attrs={"class":"panel-body text-center"})
```

```python
Nomes =[]
Escudo = []
for time in times:
 nome = time.find('span', attrs={"style":"margin-top: 10px"})
 Nomes.append(nome.get_text())
 escudo = time.find('img', attrs={"class":"icon escudo x45"})
 Escudo.append(escudo.get('src'))
```

```python
df_times = pd.DataFrame(list(zip(Escudo, Nomes)), columns=['Escudo', 'Clube'])
df_times
```

```python
df_times[['clube','Estado']] = df_times['Clube'].str.split('-', expand=True,n=1)
df_times.drop(['clube'], axis=1, inplace=True)
df_times['Sigla'] = df_times['Clube'].str[0:3]
df_times['Sigla'] = df_times['Sigla'].str.upper()
df_times['Sigla'] = df_times['Sigla'].apply(lambda a: 'AGO' if a =='ATL' else  a )
df_times['Sigla'] = df_times['Sigla'].apply(lambda a: 'ECJ' if a =='JUV' else  a )
df_times
```

```python
urljogos = 'https://www.cbf.com.br/futebol-brasileiro/jogosdehoje/campeonato-brasileiro-serie-b'
resjogos = requests.get(urljogos)
content_jogos = resjogos.content
```

```python
soup_jogos = BeautifulSoup(content_jogos, "html.parser")
jogos = soup_jogos.find_all('div',attrs={"class":"box p-t-25 p-b-15"})
```

```python
casa =[]
fora = []
escudo_casa = []
escudo_fora= []
hora =  []
for jogo in jogos:
 home = jogo.find('b', attrs={"class":"text-2 pull-right p-t-15 p-r-10 hidden-lg hidden-md"})
 casa.append(home.get_text())
 away = jogo.find('b', attrs={"class":"text-2 pull-left p-t-15 p-l-10 hidden-lg hidden-md"})
 fora.append(away.get_text())
 esc_home = jogo.find('img', attrs={"class":"icon escudo x54 pull-right"})
 escudo_casa.append(esc_home.get('src'))
 esc_away = jogo.find('img', attrs={"class":"icon escudo x54 pull-left"})
 escudo_fora.append(esc_away.get('src'))
 horario = jogo.find('div', attrs={"class":"text-3 p-t-15"})
 hora.append(horario.get_text())
```

```python
df_jogos =  pd.DataFrame(list(zip(casa, escudo_casa, hora,escudo_fora, fora)), columns=['Mandante', 'img_mandante', 'Horário','img_vis','Visitante'])
df_jogos
```

```python

tabelão = pd.merge(df_tabela_final,df_times)
tabelão = tabelão[['Pos','Evolução','Escudo','Clube','PTS','J','V','E','D','GP','GC','SG','CA','CV','Recentes','Próx']]
tabelão.drop(['Próx'],axis=1, inplace=True)
tabelão
```
